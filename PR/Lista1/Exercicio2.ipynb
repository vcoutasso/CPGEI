2. Em um problema com três classes (bidimensional), os vetores de característica em cada classe
são normalmente distribuídos de acordo com a seguinte matriz de covariância:

$$\sum = 
\begin{bmatrix}
    1.2 & 0.4 \\
    0.4 & 1.8
\end{bmatrix}$$

Os vetores de média de cada classe são $[0.5, 0.5]^T$, $[2.0, 2.0]^T$, $[1.3, 1.8]^T$.

a) Escreva uma função para plotar as três classes (usando cores diferentes), com um número
de exemplos por classe a sua escolha.

```python
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt

def plot():
    mu  = np.array([[0.5, 0.5], [2.0, 2.0], [1.3, 1.8]])
    cov = np.matrix([[1.2, 0.4], [0.4, 1.8]])
    
    for i, m in enumerate(mu):
        # Usar a operação de transposição nesse caso não tem efeito, é tudo interpretado da mesma forma
        x = np.random.multivariate_normal(m.T, cov, np.random.randint(1500, 2000))
        plt.scatter(x[:, 0], x[:, 1], s=8, label='Classe {}'.format(i+1))
    plt.legend()

if __name__ == '__main__':
    plot()
```

b) **Como você faria para construir um classificador capaz de separar essas classes (intuitivamente)?**

Um classificador capaz de separar essas classes deve ser capaz de delimitar, para cada classe, regiões no espaço de alta probabilidade de ocorrência de suas respectivas instâncias de dados. Visualmente, podemos perceber que esse exemplo traz classes de difícil separabilidade, portanto aplicar algum tipo de transformação para obter uma representação diferente pode ajudar com a tarefa.

c) **Qual a relação dos parâmetros (média e covariância) com o desempenho desse classificador?**

Médias com uma disparidade suficientemente grande entre si podem implicar em uma tarefa de classificação trivial caso seja possível obter boa separabilidade no espaço de features. Entretanto, a covariância também tem papel fundamental no que diz respeito à boa separação das classes, pois esse parâmetro é capaz de afastar instâncias de dados do ponto médio da classe, derrotando o propósito de ter pontos médios distantes um do outro.
